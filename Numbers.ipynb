```python
import pandas as pd

df = pd.read_csv('PCB Electric.csv')

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp',inplace = True)

for c in df.columns:
    df[c] = df[c].str.extract('(\d+)').astype(float)
    

df.head()
```

```python
df.describe()
```

```python
df.plot(subplots=True,figsize=(12,7))
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 96
pavg = np.zeros(n)
for i in range(96):
    pavg[i] = np.mean(df['PCB PCB-Virtual Power'].iloc[i::96])

plt.plot(pavg)
```
