```python
# Example Data arrays
d = pd.read_csv('PCB Electric Interval Data 2019.csv')

#turning the string values into floats
PCBM = d['PCB PCBM Power'].str.replace('[a-zA-Z]+', '', regex=True).astype(float)
DCB = d['PCB DCB-Main Power'].str.replace('[a-zA-Z]+', '', regex=True).astype(float)
```

```python
#The example csv has over 30 thousand data points
a = len(PCBM)
print(a)
```

```python
#Since I tried graphing every data point, it lagged my laptop out. So I'm just graphing the first thousand points:
upper = PCBM[:1000]
lower = DCB[:1000]

# turning the 15min intervals into hours for the x axis of the graph
n = len(upper)
h = int(n/4)
time = np.linspace(1, h, n)
```

```python
#vertical line connecting the upper and lower values:
plt.plot([time, time], [upper, lower], color = 'violet')

#upper and lower values:
plt.plot(time, upper, 'b.', label = 'upper value')
plt.plot(time, lower, 'r.', label = 'lower value')

#Other parts of the graph:
plt.grid()
plt.xlabel('Timestamp (h)')
plt.ylabel('Power (kW)')
plt.title('Power vs Time')
plt.legend(loc = 'upper right')
```
