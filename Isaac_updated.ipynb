```python
import pandas as pd
import numpy as np
import datetime

df = pd.read_csv('PCB Electric Interval Data 2019.csv')

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)

for c in df.columns:
    df[c] = df[c].str.extract(r'(\d+\.\d+)').astype(float)

df.dropna(inplace=True)
```

```python
time_diff = df.index[1].minute-df.index[0].minute

interval = 24*7*(60/time_diff)

min_month = df.index.month.min()
max_month = df.index.month.max()

data = []


for i in range(max_month - min_month + 1):
    md = df[df.index.month == i + 1]
    mdi = md.index
    
    # Resample the data to the calculated interval
    resampled_md = md.resample(f'{time_diff}T').mean()
    
    
    for j in range(int(interval)):
        time = mdi[j].time()
        month = mdi[j].month
        day_of_week = mdi[j].weekday() + 1
        kW = np.mean(resampled_md['Power (kW)'].iloc[j::int(interval)])
        kW_max = md['Power (kW)'].iloc[j::int(interval)].max()
        kW_min = md['Power (kW)'].iloc[j::int(interval)].min()
        print(f"Time: {time}, Month: {month}, Day of Week: {day_of_week}, kW: {kW}, Max: {kW_max}, Min: {kW_min}")

        
        data.append([time, month, day_of_week, kW])

daily_profile = pd.DataFrame(data, columns=['Time', 'Month', 'Day of Week', 'kW'])
dp = daily_profile
```
